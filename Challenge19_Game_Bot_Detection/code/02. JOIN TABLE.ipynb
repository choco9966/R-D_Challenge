```python
import pandas as pd
import numpy as np

import warnings
from glob import glob
import os

warnings.filterwarnings('ignore')

pd.set_option('max_columns',500)
```

### network

```python
[fn for fn in glob('*.csv') if 'network' in fn.lower()]
```

```python
# Train-network
train_network1 = pd.read_csv('train_network1.csv').drop('class',1)
train_network2 = pd.read_csv('train_network2.csv').drop('class',1)
train_network = pd.merge(train_network1, train_network2, on='account', how='inner')
# Test-Network
test_network1 = pd.read_csv('test_network1.csv')
test_network2 = pd.read_csv('test_network2.csv')
test_network = pd.merge(test_network1, test_network2, on='account', how='inner')
```

### player

```python
[fn for fn in glob('*.csv') if 'player' in fn.lower()]
```

```python
# Train-player
train_player = pd.read_csv('train_player.csv')

# Test-player
test_player = pd.read_csv('test_player.csv')
```

### Social

```python
[fn for fn in glob('*.csv') if 'social' in fn.lower()]
```

```python
# Train-Social
train_social = pd.read_csv('train_social.csv')
# Test-Social
test_social  = pd.read_csv('test_social.csv')
```

### Group

```python
# Train-Group
train_group  = pd.read_csv('train_group.csv')
# Test-Group
test_group  = pd.read_csv('test_group.csv')
```

###  Action

```python
# Train-Action
train_action1 = pd.read_csv('train_action1.csv').drop('class',1)
train_action2 = pd.read_csv('train_action2.csv')
train_action3 = pd.read_csv('train_action3.csv')

train_action = pd.merge( train_action1, train_action2, on='account', how='inner' )
train_action = pd.merge( train_action, train_action3, on='account', how='inner' )

# Test-Action
test_action1 = pd.read_csv('test_action1.csv')
test_action2 = pd.read_csv('test_action2.csv')
test_action3 = pd.read_csv('test_action3.csv')

test_action = pd.merge( test_action1, test_action2, on='account', how='inner' )
test_action = pd.merge( test_action, test_action3, on='account', how='inner' )
```

```python
train_action
```

### Merge

```python
train = pd.merge( train_player , train_social, on='account', how='inner' )
train = pd.merge( train, train_group , on='account', how='inner')
train = pd.merge( train, train_action, on='account', how='inner')

test = pd.merge( test_player , test_social, on='account', how='inner' )
test = pd.merge( test, test_group , on='account', how='inner')
test = pd.merge( test, test_action, on='account', how='inner')
```

## Save

```python
train.to_pickle('train.pickle')
test.to_pickle('test.pickle')
```
